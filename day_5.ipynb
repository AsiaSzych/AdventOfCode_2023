```python
import pandas as pd
from concurrent.futures import ThreadPoolExecutor
```

```python
with open('data_day5.txt') as f:
    lines = f.readlines()
lines = [line.replace("\n", "") for line in lines if line!="\n"]
```

# Part 1

```python
seeds=[int(number) for number in lines[0].split(":")[1].split(" ") if number != '']
```

```python
names = {}
current_key = ''
for line in lines[1:]:
    elems = line.split(" ")
    if len(elems) == 2:
        current_key = elems[0]
        if current_key not in names.keys():
            names[current_key] = []
        else:
            print(f"Key {current_key} alredy in dict - check out!")
    elif len(elems) == 3:
        names[current_key].append([int(el) for el in elems])
for name in names.keys():
    names[name] = sorted(names[name])
names
```

```python
df_map = pd.DataFrame( columns = ['seeds']+list(names.keys()))
df_map['seeds'] = seeds
```

```python
for i in range(len(df_map)):
    columns = df_map.columns
    for col_i in range(len(columns)-1):
        current_no = df_map.at[i, columns[col_i]]
        current_map = names[columns[col_i+1]]
        found = False
        for val in current_map:
            if val[1]<= current_no and current_no <= val[1]+val[2]:
                found = True
                item = val[0] + (current_no - val[1])
                df_map.at[i, columns[col_i+1]] = item
                break
        if not found:
            df_map.at[i, columns[col_i+1]] = current_no
```

# Part 2

```python
init_seeds=[int(number) for number in lines[0].split(":")[1].split(" ") if number != '']
seeds_pairs = [(seeds[i], seeds[i+1]) for i in range(0, len(init_seeds), 2)]
columns = ['seeds']+list(names.keys())
columns = [columns for i in range(len(seeds_pairs))]
```

```python
def sequence(start, stop):
    num = start
    while num <= stop:
        yield num
        num += 1
```

```python
def find_pair_minimum(pair, columns):
    lowest_location = -1
    # print(pair[0], pair[0]+pair[1]-1)
    seeds = sequence(pair[0], pair[0]+pair[1]-1)
    current_step = pair[0]
    while current_step < pair[0]+pair[1]-1:
        current_step = next(seeds)
        current_no = current_step
        # print(current_no)
        for col_i in range(len(columns)-1):
            current_map = names[columns[col_i+1]]
            for val in current_map:
                if val[1]<= current_no and current_no <= val[1]+val[2]:
                    item = val[0] + (current_no - val[1])
                    current_no = item
                    break
        if lowest_location == -1:
            lowest_location = current_no
        elif current_no < lowest_location:
            lowest_location = current_no
        # print(f"Exit {current_no}")
    return lowest_location
```

```python
pool = ThreadPoolExecutor(max_workers=len(seeds_pairs))
lowest = []
for result in pool.map(find_pair_minimum, seeds_pairs, columns):
    print(result)
    lowest.append(result)
```
