```python
with open('data_day15.txt') as f:
    lines = f.readlines()
line = lines[0]
```

```python
words = line.split(",")
words = [word.replace("\n", '') for word in words]
words
```

```python
def hash(word):
    cv = 0
    for litter in word:
        cv = cv + ord(litter)
        cv = cv * 17
        cv = cv % 256
    return cv 
```

## Part 1

```python
values = []
for word in words:
    cv = hash(word)
    values.append(cv)
```

```python
sum(values)
```

## Part 2

```python
keys = list(range(256))
boxes = {key: [] for key in keys}
```

```python
for word in words:
    if '-' in word:
        elems = word.split("-")
        box = hash(elems[0])
        if elems[0] in [v[0] for v in boxes[box]]:
            for val in boxes[box]:
                if val[0] == elems[0]:
                    rm = val
                    break
            boxes[box].remove(rm)
    else:
        elems = word.split("=")
        box = hash(elems[0])
        if elems[0] in [v[0] for v in boxes[box]]:
            for val in boxes[box]:
                if val[0] == elems[0]:
                    val[1] = elems[1]
                    break
        else:
            boxes[box].append([elems[0], elems[1]])
```

```python
powers = []
for box in boxes.keys():
    box_id = box+1
    lenses = boxes[box]
    for i in range(len(lenses)):
        len_power = int(lenses[i][1])
        value = box_id*(i+1)*len_power
        powers.append(value)
```

```python
sum(powers)
```
